```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, AutoModelForQuestionAnswering, AutoTokenizer, pipeline

# Load pre-trained T5 model and tokenizer for question generation
question_generator_tokenizer = T5Tokenizer.from_pretrained('valhalla/t5-base-e2e-qg')
question_generator_model = T5ForConditionalGeneration.from_pretrained('valhalla/t5-base-e2e-qg')

# Load pre-trained BERT model and tokenizer for question answering
qa_tokenizer = AutoTokenizer.from_pretrained('deepset/roberta-base-squad2')
qa_model = AutoModelForQuestionAnswering.from_pretrained('deepset/roberta-base-squad2')
qa_pipeline = pipeline('question-answering', model=qa_model, tokenizer=qa_tokenizer)
```
